### 1. set the Mega data path to be converted

### 2. Take a Look at this Mega Data

```python
m20_df = pd.read_csv(m20_path)
print(m20_df)
```

### 3.Create a CSV header for the converted file

```python
header = ""
for i in range(160*56):
    header += f"Pixel_{i},"
header += "vertex_x,vertex_y,vertex_r,vertex_phi,event"
print(header)
```

### 4. Translation Loop

```python

Sample = np.zeros((len(m20_df),56*160 + 5),dtype = object)
Max_wire_per_Superlayer = [160,160,192,224,256,288,320,352,384]
Segmentwidth_per_Superlayer = [5,5,6,7,8,9,10,11,12]

for index,each_sample in enumerate( m20_df.to_numpy() ):
    
    # Get hits and vertexs information for each coloum
    hits = each_sample[:14336]
    vertex = each_sample[-5:]
   
    # Process the hits information to hit image
    hits_wireID_list = np.where(hits == 1)[0]
    #print(hits_wireID) 
    SubSample_hit_image = np.zeros((56,160))
    for each_wireID in hits_wireID_list:
        IClayer,Iwire = BelleII.BelleCdcWire._calculateLayerWire(each_wireID)
        Superlayer = BelleII.BelleCdcWire._calculateSuperlayer_from_IClayer(IClayer)
            
        #Calculate the transformation postion 
        Segment_width_before = Segmentwidth_per_Superlayer[Superlayer]
        SegmentID = int(Iwire / Segment_width_before)
        
        Pos_in_Segment_before = Iwire % Segment_width_before
        Pos_in_Segment_after = Hits_trans_map[IClayer][Pos_in_Segment_before]
        
        Pos_X_in_Hitimage = SegmentID * 5 + Pos_in_Segment_after
        Pos_Y_in_Hitimage = IClayer
        
        SubSample_hit_image[Pos_Y_in_Hitimage][Pos_X_in_Hitimage] = 1
        #temp = SubSample_hit_image.reshape(56*160)
    #plt.imshow(SubSample_hit_image)
    #plt.show()
    SubSample_hit_image = SubSample_hit_image.reshape(56*160)
    Sample[index][:56*160] = SubSample_hit_image
    Sample[index][-5:] = vertex
    print(f"{index} Event has been processed")
np.savetxt("fuck23.csv",Sample,fmt='%s',header = header,delimiter = ",")    
 
    
```
